## Aufgabe 1

In dieser ersten Aufgabe sollen Sie Matrizen mit Hilfe von Schleifen erzeugen. Gegeben ist eine mathematische Beschreibung der Elemente der Matrix, welche mit Python umgesetzt werden soll.

Beispiel: M sei eine 5x5-Matrix mit

$$m_{i,j} = \left\{\begin{array}{cc}1&\text{wenn }i=j\\i\cdot j&\text{sonst}\end{array}\right.$$

Hier wird zuerst eine Matrix in der gewünschten Dimension mit lauter Nullen erzeugt:

```python
import numpy as np
M = np.zeros((5,5))
```

Nun wird mit zwei for-Schleifen über alle Elemente der Matrix iteriert und je nach Bedingung der entsprechende Wert gesetzt:

```python
for z in range(0,5):
    for s in range(0,5):
        if z==s:
            M[z,s] = 1
        else:
            M[z,s] = (z+1)*(s+1)
print(M)
```

**Beachten Sie:**
In der Vorlesung beginnen wir mit dem Zählen von Zeilen und Spalten einer Matrix bei 1. Die untenstehenden Definitionen sind aus der Sichtweise der Vorlesung zu verstehen. Python beginnt jedoch mit Zählen bei 0, achten Sie also darauf, dass Sie Ihre `for`-Loops entsprechend korrigieren.

Erzeugen Sie nun die folgenden Matrizen:

### Teilaufgabe a) 

A sei eine 2x10-Matrix mit

$$a_{ij} = \left\{\begin{array}{cc}i&\text{wenn }i=1\\i^2&\text{wenn }i=2\end{array}\right.$$

```python
import numpy as np
m = 2
n = 10
A = np.zeros((m, n))
print(A)
for i in np.arange(len(A)):
    for j in np.arange(len(A[i])):
        if i==0:
            A[i, j] += (i+1)
        else:
            A[i, j] += (i+1)*(i+1)
print(A)
```

### Teilaufgabe b) 

B sei eine 6x6-Matrix mit

$$b_{ij} = \left\{\begin{array}{cc}0&\text{wenn }i=j\\i+j&\text{wenn }i<j\\i-j&\text{wenn }i>j\end{array}\right.$$

```python
import numpy as np


m = 6
n = 6
B = np.zeros((m, n))

for i in range(len(B)):
    for j in range(len(B[i])):
        if(i < j):
            B[i, j] = (i+1)+(j+1)
        if(i > j):
            B[i, j] = (i+1)-(j+1)
            
print(B)
```

### Teilaufgabe c) 

C sei eine 5x5-Matrix mit

$$c_{ij} = \left\{\begin{array}{cc}0&\text{wenn }i+j<6\\1&\text{wenn }i+j=6\\2&\text{wenn }i+j>6\end{array}\right.$$

```python
import numpy as np

m = 5
n = 5

barrier = 6

C = np.zeros((m, n))

for i in range(len(C)):
    for j in range(len(C[i])):
        if(i + j > barrier):
            C[i, j] = 2
        if(i + j == barrier):
            C[i, j] = 1
            
print(C)
```

### Teilaufgabe a) 

Definieren Sie die Matrizen Z, D, B und N mit den Umsatzzahlen gemäss obigen Tabellen!


```python

# 1) Open Python shell in the same location like the script
# 2) >>> import prepare_jobs
# 3) >>> data = prepare_jobs.rewrite()
# 4) Access M by 'prepare_jobs.M'

import numpy as np

def rewrite():
    f = open('shops.dat', 'r')
    d = f.readlines()
    
    cities = ['Z', 'D', 'B', 'N']
    
    k = []
    
    cnt = 0
    
    for i in d:
        if(len(i.split('&')) > 1): 
            l = []
    
            # Insert the city label using a cyclic counter
            l.append(cities[cnt // 4])
    
            # Return a new list with sales Q1, Q2, Q3
            l = l + i.split('&')[0:4]
    
            # Append sales Q4 (split trailing slashes)
            l.append(i.split('&')[4].split('\\')[0])
    
            # Prepare writeable list
            k.append(l)
    
            # Cycle city counter
            cnt += 1
    
    output = open('output.dat', 'w')
    output.write('City Sport Q1 Q2 Q3 Q4\n')
    for i in k:
        output.write(" ".join(i) + '\n')
    output.close()
    
    filehandle = open('output.dat', 'r')
    data = filehandle.readlines()
    # for i in data:
    #     print(i, end='')
    
    return data

data = rewrite()
M = []

# Skip header line
for i in data[1:]:
    l = []
    l = i.split(' ')[2:5]
    l.append(i.split(' ')[5].split('\n')[0])
    M.append(l)

M = np.array(M)


Z = M[0:4]
D = M[4:8]
B = M[8:12]
N = M[12:16]

def convert_to_float(B):
    
    # Convert every item of the list to float
    return np.array([list(map(lambda x: float(x), Bi)) for Bi in B])

Z = convert_to_float(Z)
D = convert_to_float(D)
B = convert_to_float(B)
N = convert_to_float(N)

print("Z:")
print(Z)

print("N:")
print(N)

```

## Aufgabe 3

### Grenzverteilung bei einem (regulären)  stochastischen Prozess (Markov-Kette)

Wir unterscheiden zwischen nassen (N) und trockenen (T) Tagen. In Windisch haben statistische Untersuchungen gezeigt, dass auf einen trockenen Tag mit 70%-iger Wahrscheinlichkeit wieder ein trockener Tag folgt und dass auf einen nassen Tag mit einer Wahrscheinlichkeit von 60% wieder ein nasser Tag folgt. Heute sei ein trockener Tag.

Der obige Sachverhalt kann als stochastischer Prozess interpretiert werden. Dabei handelt es sich um einen sogenannten *Markov-Prozess*, der eine Abfolge von Zuständen erzeugt, eine sogenannte *Markov-Kette*. Die folgende Abbildung illustriert den Markov-Prozess mit Angabe der jeweiligen Übergangswahrscheinlichkeiten $P$:

Gehen wir also davon aus, dass der heutige Tag trocken ist, so ergibt sich folgender Wahrscheinlichkeitsbaum:

Um die Wahrscheinlichkeit zu berechnen, dass es nach $k$ Tagen trocken oder nass ist, müssen
wir alle Äste des Baums mit Endzustand ’T’ oder ’N’ aufsummieren, wobei wir die Wahrscheinlichkeiten entlang eines Astes multiplizieren. Wir machen das hier schnell für den abgebildeten Wahrscheinlichkeitsbaum und geben mit p(T) und p(N) die Wahrscheinlichkeiten an, dass es an diesem Tag trocken respektive nass ist:

$$
\begin{array}{ll} 
\text{Tag} & \text{Wahrscheinlichkeiten} \\ \hline
0 & p(T) = 1, p(N) = 0 \\ \hline
1 & p(T) = 0.7, p(N) = 0.3 \\ \hline
2 & {p(T) = 0.7^2 + 0.3 \cdot 0.4 = 0.61 \\ p(N) = 0.7 \cdot 0.3 + 0.3 \cdot 0.6 = 0.39 } \\ \hline
\vdots & \vdots \\ \hline
\end{array}
$$

Diese Berechnungen lassen sich dank folgender mathematischer Formulierung mittels einer Matrixmultiplikation durchführen:

Zustände:

$I = \left\{\begin{array}{cc}T&N\end{array}\right\}$

Übergangsmatrix:

$U = \left(\begin{array}{cc} p_{11} & p_{12}\\p_{21} & p_{22} \end{array}\right) = \left(\begin{array}{cc}0.7&0.4\\0.3&0.6\end{array}\right)$
     
Startverteilung (trockener Tag):

$\pi_0 = \left(\begin{array}{c}t_0\\n_0\end{array}\right) = \left(\begin{array}{c}1\\0\end{array}\right)$

Die Wahrscheinlichkeit, dass nach einem Tag ein trockener bzw. ein nasser Tag vorliegt, ist gleich:

$$\begin{eqnarray*}t_1&=&0.7\cdot t_0+0.4\cdot n_0,\\n_1&=&0.3\cdot t_0+0.6\cdot n_0,\end{eqnarray*}$$

In Matrixnotation ist dies gleich:

$$\pi_1=\left(\begin{array}{c}t_1\\n_1\end{array}\right)=\left(\begin{array}{cc}0.7&0.4\\0.3&0.6\end{array}\right)\cdot\left(\begin{array}{c}t_0\\n_0\end{array}\right)=U\cdot \pi_0$$

Also mit $t_0=1$ und $n_0=0$:

$$\pi_1 = U \cdot \pi_0 = \left(\begin{array}{cc}0.7&0.4\\0.3&0.6\end{array}\right) \cdot \left(\begin{array}{c}1\\0\end{array}\right) = \left(\begin{array}{c}0.7\\0.3\end{array}\right),$$

wie in der obigen Tabelle.

In der folgenden Zelle werden $U$ und $\pi_0$ bereits für Sie definiert:

```python
U = np.array([[0.7,0.4],[0.3,0.6]])
pi_0 = np.array([[1],[0]])
```

### Teilaufgabe c)

Erstellen Sie je einen Vektor der Wahrscheinlichkeiten für trockene bzw. nasse Tage 

$$\begin{eqnarray*}v_T &=& \left[t_0,t_1,t_2,\ldots,t_n\right]\\v_N &=& \left[n_0,n_1,n_2,\ldots,n_n\right]\end{eqnarray*}$$

für $n=30$ Tage. Stellen Sie die gefundenen Wahrscheinlichkeiten grafisch dar! Einen Teil des Codes haben wir Ihnen bereits vorgegeben:


```python
n = 30
pi_k = pi_0
v_T = np.zeros(n)
v_N = np.zeros(n)

for k in range(n):
    v_T[k] = np.dot(U, pi_k)[0]
    v_N[k] = np.dot(U, pi_k)[1]
    pi_k = np.dot(U, pi_k)

print( "v_T = {}".format( v_T ) )
print( "v_N = {}".format( v_N ) )
    
import matplotlib.pyplot as plt
x = np.arange(0, n, 1)
plt.figure( figsize=(10,7) )
plt.plot( x, v_T, color="red", label="p(T)" )
plt.plot( x, v_N, color="blue", label="p(N)" )
plt.xlabel("Tag")
plt.legend()
print(plt.show())
```
